```python
ogdf = pd.read_csv('Car_sales_dataset_raw.csv')
spdf = pd.read_csv('Salesperson_table.csv')
cdf = pd.read_csv('Customer_table.csv')
cardf = pd.read_csv('Car_table.csv')
ogdf.sample(5)
            
```

```python
ogdf.columns
```

```python
ogdf['Car_Make_Model'] = ogdf['Car Make'] + ' '+ ogdf['Car Model']
car_make_model = ogdf.pop('Car_Make_Model')
ogdf.insert(7,'Car_Make_Model',car_make_model)
ogdf
```

```python
ogdf['Salesperson'] = ogdf['Salesperson'].str.replace(r'^(Mr\.|Mrs\.|Ms\.|Dr\.)\s+', '', regex=True)
ogdf['Salesperson'] = ogdf['Salesperson'].str.replace(
    r'[, ]*\b(?:M\.?D\.?|D\.?D\.?S\.?|D\.?V\.?M\.?|P(?:H|\.H\.?)\.?D\.?)\b[, ]*$',
    '',
    regex=True,
).str.strip()
```

```python
ogdf_flag = pd.merge(ogdf,spdf, left_on = 'Salesperson', right_on = 'Sales person', how = 'left' )
ogdf_spid = ogdf_flag[['Salesperson_id']]
ogdf_spid
```

```python
ogdf_spid.isnull().sum()
```

```python
ogdf.columns
```

```python
ogdf['Salesperson_id'] = ogdf_spid
sh = ogdf.pop('Salesperson_id')
ogdf.insert(1,'Salesperson_id', sh)
print(ogdf.columns)
del ogdf['Salesperson']
```

```python
ogdf.columns
```

```python
import re

ogdf['Customer Name'] = ogdf['Customer Name'].str.replace(r'^(Mr\.|Mrs\.|Ms\.|Dr\.)\s+', '', regex=True)
ogdf['Customer Name'] = ogdf['Customer Name'].str.replace(
    r'[, ]*\b(?:M\.?D\.?|D\.?D\.?S\.?|D\.?V\.?M\.?|P(?:H|\.H\.?)?\.?D\.?)\b[, ]*$', 
    '', 
    regex=True,
    flags=re.IGNORECASE  # Important for case-insensitive matching
).str.strip()
```

```python
ff = pd.merge(ogdf,cdf,left_on = 'Customer Name', right_on = 'Customer Name', how = 'left')
cust_id = ff['Customer ID']
```

```python
ogdf.insert(2,'Customer_id', cust_id)
ogdf
```

```python
del ogdf['Customer Name']
del ogdf['Customer Age']
del ogdf['Customer Gender']
ogdf
```

```python
car_merge = pd.merge(ogdf,cardf,left_on = 'Car_Make_Model', right_on = 'Car_Make_model', how = 'left')
carid = car_merge['Car_id']
carid
```

```python
ogdf.insert(3,"Car_id", carid)
del ogdf['Car Make']
del ogdf['Car Model']
del ogdf['Car_Make_Model']
ogdf
```
