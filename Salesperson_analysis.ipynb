```python
import pandas as pd
import numpy as np
import re
```

```python
og_df = pd.read_csv("Car_sales_dataset_raw.csv")
og_df
```

```python
og_df['Salesperson'] = og_df['Salesperson'].str.replace(r'^(Mr\.|Mrs\.|Ms\.|Dr\.)\s+', '', regex=True)
og_df['Salesperson'] = og_df['Salesperson'].str.replace(
    r'[, ]*\b(?:M\.?D\.?|D\.?D\.?S\.?|D\.?V\.?M\.?|P(?:H|\.H\.?)\.?D\.?)\b[, ]*$',
    '',
    regex=True,
    og_df=re.IGNORECASE
).str.strip()


sp = og_df['Salesperson'].unique()
sp
```

```python
spn = pd.DataFrame(sp, columns = ['Sales person'])
spn
```

```python
sp_id = np. linspace (101,201,100, dtype= int)
print(sp_id)
```

```python
spn['Salesperson_id'] = sp_id
spn
```

```python
flag = spn.pop('Salesperson_id')
spn.insert(0,'Salesperson_id' ,flag)
spn
```

```python
spn.to_csv("Salesperson_table.csv", index = False)
#Prefix and suffix removed 
```
