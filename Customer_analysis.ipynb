```python
import numpy as np
import pandas as pd
import re
#columns - customer_id ,  name,  age, gender
```

```python
og_df = pd.read_csv("Car_sales_dataset_raw.csv")
og_df.sample(10)
```

```python
og_df['Customer Name'] = og_df['Customer Name'].str.replace(r'^(Mr\.|Mrs\.|Ms\.|Dr\.)\s+', '', regex=True)
og_df['Customer Name'] = og_df['Customer Name'].str.replace(
    r'[, ]*\b(?:M\.?D\.?|D\.?D\.?S\.?|D\.?V\.?M\.?|P(?:H|\.H\.?)\.?D\.?)\b[, ]*$',
    '',
    regex=True,
    flags=re.IGNORECASE
).str.strip()
og_df['Customer Name']

#Removes unwanted prefix and suffix #DVM PHD 
```

```python
ww = og_df.drop_duplicates(subset = ['Customer Name'])
uc = ww['Customer Name']
ua = ww['Customer Age']
ug = ww['Customer Gender']
ww[['Customer Name', 'Customer Age', 'Customer Gender']]
```

```python
cid = np.arange(1,301587,step = 1, dtype = int)

uc = ww['Customer Name'].unique()
c_a = pd.DataFrame({'Customer ID': cid, 'Customer Name': uc, 'Customer Age': ua, 'Customer Gender':ug})
c_a

#After removing the prefix and suffix, almost 32000 duplicate rows were reduced
```

```python
c_a.to_csv('Customer_table.csv', index= False)
```
